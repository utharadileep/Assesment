Formulate questions based on the ny_flights dataset

What is the average departure delay across all flights?

Which airport (JFK, LGA, EWR) had the most departures?

Multivariate Analysis:
Investigate interactions among multiple variables (e.g., pair plots, 3D plots)

````python
sns.pairplot(df1, diag_kind='kde', markers='+')
plt.suptitle('Pair Plots of All Variables', y=1.02)
plt.show()
````

````python
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
x = df1['Glucose']
y = df1['BMI']
z = df1['Age']
ax.scatter(x, y, z, c='r', marker='o')

ax.set_xlabel('Glucose')
ax.set_ylabel('BMI')
ax.set_zlabel('Age')
ax.set_title('3D Scatter Plot of Glucose, BMI, and Age')

plt.show()
````
